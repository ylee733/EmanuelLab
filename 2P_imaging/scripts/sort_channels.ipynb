```python
import os
import shutil
```

```python
TSeries = r"N:\Lee_Yuna\projects\2P_imaging\BPT2\20251013\TSeries-10132025-0917-041"
```

```python
ch1_folder = os.path.join(TSeries, 'Ch1')
ch2_folder = os.path.join(TSeries, 'Ch2')
os.makedirs(ch1_folder, exist_ok=True)
os.makedirs(ch2_folder, exist_ok=True)
```

```python
for fname in os.listdir(TSeries):
    if not fname.lower().endswith('.tif'):
        continue  # skip non-TIFF files
    full_path = os.path.join(TSeries, fname)
    if '_Ch1' in fname:
        shutil.move(full_path, os.path.join(ch1_folder, fname))
    elif '_Ch2' in fname:
        shutil.move(full_path, os.path.join(ch2_folder, fname))

print("done sorting")
```
