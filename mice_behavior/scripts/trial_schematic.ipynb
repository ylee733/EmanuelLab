```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(8, 2))

# Timeline range
ax.hlines(1, -1, 7, color='black')

# Stimulus
ax.hlines(1, 1, 2, color='blue', linewidth=8, label='Stimulus')
ax.text(1.5, 1.1, 'Stimulus\n1–2 s', ha='center')

# Response window
ax.hlines(1, 2, 2.5, color='green', linewidth=8, label='Response Window')
ax.text(2.25, 1.1, 'Response\n2–2.5 s', ha='center')

# No lick
ax.hlines(1, -1, 0, color='red', linewidth=8, label='Response Window')
ax.text(-0.3, 1.1, 'No Lick Period\n-1–0 s', ha='center')

# annotation at x=0
ax.text(0, 0.7, "Start of Trial", ha='center', va='top')

# Formatting
ax.set_ylim(0.5, 1.5)
ax.set_yticks([])
ax.set_xlim(-1, 7)
ax.set_xlabel('Time (s)')
ax.set_title('Trial Structure')

plt.tight_layout()
plt.show()
```
